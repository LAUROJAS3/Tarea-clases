```python
 #Creamos una clase
class Gasolinera:
 
#Creamos y llamamos a los metodos
    def __init__(self):
       self.costo1=20.25
       self.costo2=21.94
       self.costo3=21.61
 
    def magna(self,l1):
        return l1*self.costo1
 
    def premium(self,l2):
        return l2*self.costo2
    
    def diesel(self,l3):
        return l3*self.costo3
 
    def cigarros(self,p1):
        return p1*5
 
    def agua(self,p2):
        return p2*10

 
#Del bloque principial creamos un objeto de la clase Gasolinera 'gas'
gas=Gasolinera()

def menu():
    opcion=0
    while opcion!=6:   
        print("Gasolinera los michis")
        print("1.Magna")
        print("2.Premium")
        print("3.Diesel")
        print("4.Otros servivios")
        print("5.Personal autorizado")
        print("6.Salir")
        opcion = int(input("Ingrese una opcion: "))
#Utilizamos una forma simplificada de if anidados con la sentencia elif
        if opcion==1:
            print("Has elegido: Magna")
            l1=float(input("Cantidad de litros:"))
            print("El total es: ",gas.magna(l1),"pesos")  
        elif opcion==2:
            print("Has elegido: Premium")
            l2=float(input("Cantidad de litros:"))
            print("El total es: ",gas.premium(l2),"pesos")   
        elif opcion==3:
            print("Has elegido: Diesel")
            l3=float(input("Cantidad de litros:"))
            print("El total es: ",gas.diesel(l3),"pesos")
#Para la opcion 4 tambien ocupamos la forma simplificada de if anidados para 
#mostrarle mas servicios al usuario
        elif opcion==4:
            print("1.Cigarrillos $5")
            print("2.Agua embotellada $10")
            op=int(input("Ingresa una opcion: "))
            if op == 1:
              print("1.Malboro")
              print("2.Camel")
              print("3.Delicados")
              op1=int(input("Ingresa una opcion: "))
              if op1==1:
                 print("Compra de cigarrillos Malboro")
                 p1=float(input("Cantidad de cigarrillos:"))
                 print("El total es: ",gas.cigarros(p1),"pesos")
                 print("Usted ha salido de Otros servicios")
              elif op1==2:
                 print("Compra de cigarrillos Camel")
                 p1=float(input("Cantidad de cigarrillos:"))
                 print("El total es: ",gas.cigarros(p1),"pesos")
                 print("Usted ha salido de Otros servicios")
              elif op1==3:
                 print("Compra de cigarrillos Delicados")
                 p1=float(input("Cantidad de cigarrillos:"))
                 print("El total es: ",gas.cigarros(p1),"pesos")
                 print("Usted ha salido de Otros servicios")
            elif op ==2:
               print("Compra de agua embotellada de 1 Lt")
               p2=float(input("Cantidad de agua:"))
               print("El total es: ",gas.agua(p2),"pesos")
               print("Usted ha salido de Otros servicios")
    
        elif opcion==5:
            clv=2258
            usu = input("Introduce la clave: ")
            usu= int (usu)
            if usu == clv:
               print("Acceso correcto")
               print("1.Actualizar precios")
               print("2.Salir")
               op2=int(input("Ingresa una opcion: "))
               if op2==1:
                  print("1.Actualizar precio de magna")
                  print("2.Actualizar precio de premium")
                  print("3.Actualizar precio de diesel")
                  op3=int(input("Ingresa una opcion: "))
                  if op3==1:
                     costo1=float(input("Nuevo precio:"))
                     print("La nueva tarifa es: ",costo1,"pesos")
                  elif op3==2:
                     costo2=float(input("Nuevo precio: "))
                     print("La nueva tarifa es: ", costo2,"pesos")
                  elif op3==3:
                     costo3=float(input("Nuevo precio: "))
                     print("La nueva tarifa es: ", costo3,"pesos")
            else:
               print("Denegar")

        elif opcion==6:
            print("Vuelva pronto")
 
menu()
```
